```python
#import необходимых модулей

import pandas as pd
from catboost import CatBoostRegressor, Pool
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score

import pickle
```

```python
#Считывание данных в DataFrame 

train = pd.read_csv('./content/train_dataset_train.csv', sep=';', index_col=None, dtype={'PATIENT_SEX':str, 'MKB_CODE':str, 'ADRES':str, 'VISIT_MONTH_YEAR':str, 'AGE_CATEGORY':str, 'PATIENT_ID_COUNT':int})
```

```python
visit_months = train['VISIT_MONTH_YEAR'].unique().tolist()
```

```python
import functools

def compare_function(vm1, vm2):
    if vm1[3:] > vm2[3:]:
        return 1
    elif vm1[3:] < vm2[3:]:
        return -1
    else:
        if vm1[:2] >= vm2[:2]:
            return 1
        else:
            return -1

visit_months.sort(key=functools.cmp_to_key(compare_function))
```

```python
loaded_model = list() # загруженные модели
for j in range(len(visit_months)):
    vmonth = visit_months[j]
    print(f"Обработка {vmonth}")    
    trainvm = train[train['VISIT_MONTH_YEAR'] == vmonth]
    trainvm = trainvm.drop('VISIT_MONTH_YEAR', axis=1)
    #Отделение меток от данных
    X = trainvm[['PATIENT_SEX', 'MKB_CODE', 'ADRES', 'AGE_CATEGORY']]
    y = trainvm[['PATIENT_ID_COUNT']]
    #Создание объекта данных Pool, плюсы: возможность указать какие признаки являются категориальными
    pool_train = Pool(X, y, cat_features = ['PATIENT_SEX', 'MKB_CODE', 'ADRES', 'AGE_CATEGORY'])
    #Объявление CatBoostRegressor и обучение
    model = CatBoostRegressor(task_type='CPU', depth=10, l2_leaf_reg=7, learning_rate=0.1 )
    model.fit(pool_train, verbose=False)
    filename = f'finalized_model{j:02}.sav'
    pickle.dump(model, open(filename, 'wb'))
    loaded_model.append(filename)
```
